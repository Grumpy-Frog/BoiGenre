```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your CSV file into a pandas DataFrame
file_path = 'temp/books_with_synopsis_unique_gen_sum.csv'
df = pd.read_csv(file_path)

# Split the data into train, test, and validation sets
train_data, temp_data = train_test_split(df, test_size=0.4, random_state=42)
test_data, valid_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# Save the resulting sets to new CSV files
train_data.to_csv('Dataset_60_20_20/train.csv', index=False)
test_data.to_csv('Dataset_60_20_20/test.csv', index=False)
valid_data.to_csv('Dataset_60_20_20/validation.csv', index=False)
```
