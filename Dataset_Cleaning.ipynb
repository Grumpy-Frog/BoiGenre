```python
import pandas as pd
import re

# Function to count sentences in Bangla
def count_sentences(text):
    if pd.isna(text):  # handle missing summaries
        return 0
    # Split on Bangla danda (), question mark, or exclamation
    sentences = re.split(r"[?!]", str(text))
    return len([s for s in sentences if s.strip()])

# Create a new column with sentence counts
df["SentenceCount"] = df["Summary"].apply(count_sentences)

# Group by Genre and get stats
sent_per_genre = (
    df.groupby("Genre")["SentenceCount"]
      .agg(["mean", "min", "max", "median", "count"])
      .sort_values("mean", ascending=False)
)

print(sent_per_genre)
```

```python
import re

# Count words in Bangla/English summaries
def count_words(text):
    if pd.isna(text):
        return 0
    # match Bangla letters + English letters/digits as tokens
    return len(re.findall(r'[\u0980-\u09FFA-Za-z0-9]+', str(text)))

# 1) Number of books (unique titles)
num_books = df['Title'].nunique()

# 2) Number of different authors
num_authors = df['Author'].nunique()

# Word-count per summary
wc = df['Summary'].apply(count_words)

# 3) Max word length (per-summary word count)
max_word_len = wc.max()

# 4) Min word length
min_word_len = wc.min()

# 5) Avg word length
avg_word_len = wc.mean()

# 6) Number of Genre (unique)
num_genres = df['Genre'].nunique()

print(f"Number of books (unique titles): {num_books}")
print(f"Number of different authors: {num_authors}")
print(f"Max summary word count: {max_word_len}")
print(f"Min summary word count: {min_word_len}")
print(f"Average summary word count: {avg_word_len:.2f}")
print(f"Number of genres: {num_genres}")
```

```python
import pandas as pd
import re

# Function to count sentences in Bangla
def count_sentences(text):
    if pd.isna(text):  # handle missing summaries
        return 0
    sentences = re.split(r"[?!]", str(text))
    return len([s for s in sentences if s.strip()])

# Function to count words
def count_words(text):
    if pd.isna(text):
        return 0
    # Split on Bangla/English word boundaries
    words = re.findall(r"[\u0980-\u09FFA-Za-z0-9]+", str(text))
    return len(words)

# Add new columns
df["SentenceCount"] = df["Summary"].apply(count_sentences)
df["WordCount"] = df["Summary"].apply(count_words)

# Group by Genre and compute stats
stats_per_genre = (
    df.groupby("Genre")
      .agg(
          Avg_Sentences=("SentenceCount", "mean"),
          Min_Sentences=("SentenceCount", "min"),
          Max_Sentences=("SentenceCount", "max"),
          Median_Sentences=("SentenceCount", "median"),
          Total_Sentences=("SentenceCount", "sum"),
          Avg_Words=("WordCount", "mean"),
          Max_Words=("WordCount", "max"),
          Min_Words=("WordCount", "min"),
          Total_Words=("WordCount", "sum"),
      )
      .sort_values("Avg_Sentences", ascending=False)
)

print(stats_per_genre)
```

```python
import pandas as pd
import numpy as np
import re

# --- Helpers ---
def count_sentences(text):
    if pd.isna(text): return 0
    parts = re.split(r"[?!]", str(text))
    return sum(1 for s in parts if s.strip())

def count_words(text):
    if pd.isna(text): return 0
    return len(re.findall(r"[\u0980-\u09FFA-Za-z0-9]+", str(text)))

# Lengths per summary
df["SentenceCount"] = df["Summary"].apply(count_sentences)
df["WordCount"]     = df["Summary"].apply(count_words)

# Per-genre core stats: N, meanstd for words and sentences
grp = df.groupby("Genre")
stats_core = pd.DataFrame({
    "N": grp.size(),
    "WordsMean": grp["WordCount"].mean(),
    "WordsStd":  grp["WordCount"].std(ddof=1),
    "SentsMean": grp["SentenceCount"].mean(),
    "SentsStd":  grp["SentenceCount"].std(ddof=1),
}).round(2)

# Optional: collapse meanstd into a single string column (nice for LaTeX)
stats_core["Words / Summ."] = stats_core.apply(
    lambda r: f"{r['WordsMean']:.2f}{r['WordsStd']:.2f}", axis=1)
stats_core["Sents / Summ."] = stats_core.apply(
    lambda r: f"{r['SentsMean']:.2f}{r['SentsStd']:.2f}", axis=1)

# Final minimal table
table_acl = stats_core[["N", "Words / Summ.", "Sents / Summ."]].sort_values("N", ascending=False)
print(table_acl)
```

# Most frequent words each genre

```python
import pandas as pd
from collections import Counter
import re

# Simple tokenizer for Bangla + English (split by spaces/punctuation)
def tokenize(text):
    if pd.isna(text):
        return []
    # keep Bangla letters, English letters, digits
    return re.findall(r'[\u0980-\u09FFa-zA-Z0-9]+', str(text))

# Function: most frequent words for one genre
def most_frequent_words(texts, top_n=15):
    all_tokens = []
    for t in texts:
        all_tokens.extend(tokenize(t))
    counter = Counter(all_tokens)
    return counter.most_common(top_n)

# Apply per Genre
top_words_per_genre = (
    df.groupby("Genre")["Summary"]
      .apply(lambda summaries: most_frequent_words(summaries, top_n=10))
)

# Convert to DataFrame for easier viewing
top_words_df = pd.DataFrame(top_words_per_genre)
print(top_words_df.head())
```

```python
import pandas as pd
from collections import Counter
import re
import unicodedata

# Count visible letters/numbers (ignores combining marks)
def ln_count(token):
    return sum(1 for ch in token if unicodedata.category(ch)[0] in ('L','N'))

# Tokenizer for Bangla + English, filter by "human" length >= 2
def tokenize(text, min_len=2):
    if pd.isna(text):
        return []
    tokens = re.findall(r'[\u0980-\u09FFA-Za-z0-9]+', str(text))
    return [tok for tok in tokens if ln_count(tok) >= min_len]

# Most frequent words for one genre
def most_frequent_words(texts, top_n=15):
    all_tokens = []
    for t in texts:
        all_tokens.extend(tokenize(t, min_len=2))
    counter = Counter(all_tokens)
    return counter.most_common(top_n)

# Apply per genre
top_words_per_genre = (
    df.groupby("Genre")["Summary"]
      .apply(lambda summaries: most_frequent_words(summaries, top_n=15))
)

# Expand into table with words + counts
def expand_top_words(row):
    data = {}
    for i, (w, c) in enumerate(row, start=1):
        data[f"Word{i}"] = w
        data[f"Freq{i}"] = c
    return pd.Series(data)

top_words_table = top_words_per_genre.apply(expand_top_words)

# Show full table
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
print(top_words_table)
```

# 1 Vocabulary Size per Genre (unique words)

```python
import re

# Tokenizer (Bangla + English)
def tokenize(text):
    if pd.isna(text):
        return []
    return re.findall(r'[\u0980-\u09FFA-Za-z0-9]+', str(text))

# Compute unique vocab per genre
vocab_per_genre = (
    df.groupby("Genre")["Summary"]
      .apply(lambda summaries: set(token for text in summaries for token in tokenize(text)))
      .apply(len)  # count unique words
      .sort_values(ascending=False)
)

print("Vocabulary size per genre:")
print(vocab_per_genre)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sort by value (already sorted in vocab_per_genre)
values = vocab_per_genre.values
labels = [f"{genre} ({count})" for genre, count in vocab_per_genre.items()]

# Colors: deep (high)  light (low)
colors = plt.cm.Blues(np.linspace(0.9, 0.3, len(values)))

# Pie chart
fig, ax = plt.subplots(figsize=(8,8))
wedges, texts = ax.pie(
    values,
    labels=labels,
    startangle=140,
    colors=colors,
    textprops={'fontsize': 10}
)

# Title
ax.set_title("Vocabulary Size Distribution by Genre", fontsize=14)

plt.tight_layout()
plt.show()
```

# 2 Overlapping Vocabulary between Genres (Cosine Similarity of TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Group all summaries by genre (concatenate)
genre_texts = df.groupby("Genre")["Summary"].apply(lambda x: " ".join(map(str, x)))

# TF-IDF representation
vectorizer = TfidfVectorizer(token_pattern=r'[\u0980-\u09FFA-Za-z0-9]+')
tfidf_matrix = vectorizer.fit_transform(genre_texts)

# Cosine similarity between genres
cosine_sim = cosine_similarity(tfidf_matrix)

# Convert to DataFrame for readability
cosine_df = pd.DataFrame(cosine_sim, index=genre_texts.index, columns=genre_texts.index)

print("Cosine similarity of vocab between genres:")
print(cosine_df.round(3))
```

# Type-Token Ratio (TTR)

 What this shows

Higher TTR  more diverse vocabulary relative to text length (e.g., Poetry).

Lower TTR  more repetitive language (e.g., Math textbooks may repeat terms).

```python
import re
import pandas as pd

# Tokenizer (Bangla + English)
def tokenize(text):
    if pd.isna(text):
        return []
    return re.findall(r'[\u0980-\u09FFA-Za-z0-9]+', str(text))

# Function to calculate TTR for a list of texts
def calc_ttr(texts):
    tokens = [tok for text in texts for tok in tokenize(text)]
    if len(tokens) == 0:
        return 0
    return len(set(tokens)) / len(tokens)

# Calculate TTR per genre
ttr_per_genre = df.groupby("Genre")["Summary"].apply(calc_ttr).sort_values(ascending=False)

print("Type-Token Ratio (TTR) per genre:")
print(ttr_per_genre)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Already computed ttr_per_genre
ttr_per_genre = df.groupby("Genre")["Summary"].apply(calc_ttr).sort_values(ascending=False)

# Bar plot
fig, ax = plt.subplots(figsize=(12,6))

# Gradient color (deep to light blue, leftright)
colors = plt.cm.Blues(np.linspace(0.9, 0.3, len(ttr_per_genre)))

x = np.arange(len(ttr_per_genre))
bars = ax.bar(x, ttr_per_genre.values, color=colors)

# Labels
ax.set_xticks(x)
ax.set_xticklabels(ttr_per_genre.index, rotation=30, ha='right')

# Add value labels
for p in bars:
    ax.text(p.get_x() + p.get_width()/2,
            p.get_height() + 0.01,
            f"{p.get_height():.2f}",
            ha='center', va='bottom', fontsize=9)

ax.set_title("Type-Token Ratio (TTR) per Genre")
ax.set_xlabel("Genre")
ax.set_ylabel("TTR (Lexical Diversity)")
ax.set_ylim(0, 0.3)   # since TTR is always 1

plt.tight_layout()
plt.show()
```
